# swap

```python
# method 1
a ,b = 5,6
temp = a
a = b
b = temp
print(a,b)
```

```python
# method 2 without 3rd variable
a, b = 4,3
a = a + b # 7
b = a - b
a = a - b
print(a,b)
```

```python
# method 3 python feature
a,b = 8,9
a,b = b,a
print(a,b)
```

# Set 
same as list but can not be duplicate and slice or query or index

can update it (used hash-table)

also auto sort

```python
set_val = {2,2,2,3,4,5,6}
set_val
```

```python
set_val[1]
```

```python
set_val1 = set([9,8,4,5])
set_val1
```

```python
print(set_val,set_val1)
#uninon 
print(set_val | set_val1)
# inserstion
print(set_val & set_val1)
# difference
print(set_val - set_val1)
# symetric difference
print(set_val ^ set_val1)
```
